```python
#Clean Data

df_cleaned = df.dropna(axis=1 , how='all').copy()
```

```python
df.shape
```

```python
df_cleaned.shape
```

```python
df_cleaned.iloc[:,4:] = df_cleaned.iloc[:, 4:].apply(lambda row: row.fillna(row.mean()), axis=1)
df_cleaned.isnull().sum().sum()
```

```python
#to handle missing values that exist in the middle years,
#I prefer imputing them rather than dropping. Since population density trends are usually stable over time,
#filling missing values with the mean of each country's available years is a reasonable approach.
```

```python
df_cleaned.iloc[:, 4:] = df_cleaned.iloc[:, 4:].fillna(df_cleaned.iloc[:, 4:].mean())
df_cleaned.isnull().sum().sum()
```

```python
#First, I imputed missing values using each country’s average, which works when a country has at least some valid data.
#For countries with all values missing in certain years, I then used the year-wise average to fill remaining NaNs.
#This two-step strategy ensures no data loss and smooths the dataset properly.
```

```python
#import continent in other file and merge ti base file
import country_converter as coco

cc = coco.CountryConverter()
df_merged=df_cleaned.copy()
df_merged['Continent'] = cc.convert(names=df['Country Name'].tolist(), to='continent')
df_merged[df_merged['Continent']=='not found']['Country Name'].unique()

df_merged= df_merged[df_merged['Continent'] != 'not found']
df_merged.head()
```

```python

columns_1961_2015 = [str(year) for year in range(1961, 2016)]

# محاسبه میانگین برای هر کشور
df_cleaned['1961_2015_avg'] = df_merged[columns_1961_2015].mean(axis=1)

# مرتب‌سازی و نمایش 10 کشور با بیشترین میانگین چگالی جمعیت
top_10_avg_density = df_merged[['Country Name','Continent', '1961_2015_avg']].sort_values(by='1961_2015_avg', ascending=False).head(10)
top_10_avg_density

```

```python
#From 2000 to 2015, Macao and Monaco had the highest average population density by far,
#with over 17,000 people per square kilometer. The rest of the top 10 includes densely
#populated regions and small island countries. Bangladesh stands out as a large, non-island country with
#an extremely high density, reflecting population pressure.
```

```python
import matplotlib.pyplot as plt

global_avg = df_merged[columns_1961_2015].mean()

plt.figure(figsize=(10,5))
plt.plot(global_avg.index, global_avg.values,marker='o')
plt.title('Global Average Population Density(1961-2015)')
plt.xlabel('Year')
plt.ylabel('Average Density')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_merged['avg_2000s']= df_merged[[str(y) for y in range(2000,2010)]].mean(axis=1)
df_merged['avg_2010s']= df_merged[[str(y) for y in range(2010,2016)]].mean(axis=1)
```

```python
pivot = df_merged[['Country Name','avg_2000s','avg_2010s']]
pivot_sorted = pivot.sort_values(by='avg_2010s',ascending=False)
pivot_sorted.head(10)
```

```python
years = [col for col in df_merged.columns if col.isdigit()]
years = sorted([int(year) for year in years if int(year)>=1961 and int(year)<=2015])

interval = [(start, start+3) for start in range(1961,2013,4)]

for start, end in interval:
    period = [str(y) for y in range(start, end+1)]
    df_merged[f'{start}_{end}_avg']=df_merged[period].mean(axis=1)
    avg_cols = [col for col in df_merged.columns if '_avg' in col]
df_avg = df_merged[['Country Name']  + avg_cols]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# برای هر قاره، انتخاب 3 کشور با بیشترین میانگین جمعیت بین 2000 تا 2015
top_3_per_continent = df_merged.sort_values('1961_2015_avg', ascending=False).groupby('Continent').head(3)

# فقط ستون‌های مورد نیاز برای نمایش
top_3_per_continent = top_3_per_continent[['Country Name', 'Continent', '1961_2015_avg']].reset_index(drop=True)

#import ace_tools as tools; tools.display_dataframe_to_user(name="Top 3 Countries per Continent by Population Density", dataframe=top_3_per_continent)


import matplotlib.pyplot as plt
import seaborn as sns

# مرتب‌سازی بر اساس قاره و جمعیت
top_3_sorted = top_3_per_continent.sort_values(['Continent', '1961_2015_avg'], ascending=[True, False])

# ترتیب محور x بر اساس کشورها مرتب‌شده
plt.figure(figsize=(14, 7))
sns.set_style("whitegrid")

sns.barplot(
    data=top_3_sorted,
    x='Country Name',
    y='1961_2015_avg',
    hue='Continent',
    dodge=False,
    palette='Set2',
    order=top_3_sorted['Country Name']  # ترتیب نمایش دقیق محور x
)

# عنوان و برچسب‌ها
plt.title('Top 3 Countries by Population Density per Continent (1961_2015_avg)', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Avg Population Density', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Continent')
plt.tight_layout()
plt.show()
```
